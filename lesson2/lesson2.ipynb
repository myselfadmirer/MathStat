## 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

### Так как дана вероятность попадания для одного выстрела, то решим с помощью биномиального распределения:
### $P=C^k_n \cdot p^k\cdot q^{n-k}=C^{85}_{100}\cdot p^{85}\cdot q^{100-85}=\frac {100!}{85!15!}\cdot0,8^{85}\cdot0,2^{15}=0,048$

## 2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

### Так как дан промежуток времени, используем закон распределения Пуассона:
### $P=\frac{\lambda^m}{m!}\cdot e^{-\lambda}$
### $\lambda=M(X)=n\cdot p=5000\cdot 0,0004=2$
### В первом случае при 0 перегоревших:
### $P=\frac{2^0}{0!}\cdot e^{-2}=0,135$
## Во втором случае при 2 перегоревших:
### $P=\frac{2^2}{2!}\cdot e^{-2}=0,27$

## 3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

### Как и в задании №1, подходит биномиальное распределение.
### $P=C^k_n \cdot p^k\cdot q^{n-k}=C^{70}_{144}\cdot p^{70}\cdot q^{144-70}=\frac {144!}{70!74!}\cdot0,5^{70}\cdot0,5^{74}=0,0688$
### С другой стороны, вычисления кажутся громоздкими, n>100, k>20, а вероятности p и q равны 0,5, следовательно, можно решить данную задачу с помощью теоремы Муавра-Лапласа:
### $x=\frac{m-np}{\sqrt{npq}}=\frac{70-144\cdot0,5}{\sqrt{144\cdot0,5\cdot0,5}}=\frac{-2}{6}=-\frac1{3}$
### $\phi(x)=\frac1{\sqrt{2\pi}}\cdot e^{-\frac{x^2}{2}}=\frac1{\sqrt{2\pi}}\cdot e^{-\frac{(-\frac1{3})^2}{2}}=0,3774$
### $P=\frac1{\sqrt{npq}}\cdot \phi(x)=\frac1{\sqrt{144\cdot0,5\cdot0,5}}\cdot 0,3774=0,0629$

## 4. В первом ящике находится 10 мячей, из которых 7 - белые. Во втором ящике - 11 мячей, из которых 9 белых. Из каждого ящика вытаскивают случайным образом по два мяча. 
### Гипергеометрическое распределение

### 1) Какова вероятность того, что все мячи белые? 
### $P(4)=\frac{C^2_7\cdot C^0_3}{C^2_{10}}\cdot \frac{C^2_9\cdot C^0_2}{C^2_{11}}=\frac{21\cdot 1}{45}\cdot \frac{36\cdot 1}{55}=0,3055$

### 2) Какова вероятность того, что ровно два мяча белые?
### $P(2)=\frac{C^2_7\cdot C^0_3}{C^2_{10}}\cdot \frac{C^0_9\cdot C^2_2}{C^2_{11}}+\frac{C^1_7\cdot C^1_3}{C^2_{10}}\cdot \frac{C^1_9\cdot C^1_2}{C^2_{11}}+\frac{C^0_7\cdot C^2_3}{C^2_{10}}\cdot \frac{C^2_9\cdot C^0_2}{C^2_{11}}=\frac{21\cdot 1}{45}\cdot \frac{1\cdot 1}{55}+\frac{7\cdot 3}{45}\cdot \frac{9\cdot 2}{55}+\frac{1\cdot 3}{45}\cdot \frac{36\cdot 1}{55}=0,2049$

### 3) Какова вероятность того, что хотя бы один мяч белый?
### $P=1-\frac{C^0_7\cdot C^2_3}{C^2_{10}}\cdot \frac{C^0_9\cdot C^2_2}{C^2_{11}}=1-\frac{1\cdot 3}{45}\cdot \frac{1\cdot 1}{55}=0,9988$
